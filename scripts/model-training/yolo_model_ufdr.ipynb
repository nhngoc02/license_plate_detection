```python
from ultralytics import YOLO
```

# Model 4 - Train from UFDR

## Train YOLO model based on UFDR

```python
model_ufdr = YOLO('yolov8n_ufdr.pt')  
model_ufdr.train(
    data='/UFPR-ALPR_dataset/ufpr_alpr_yolo.yaml',  
    epochs=5,
    imgsz=640,
    batch=16,
    name='yolov8_ufdr'
)
```

## Test model trained on UFDR

```python
model_ufdr.val(data='UFPR-ALPR_dataset/ufpr_alpr_yolo.yaml', split='test')
```

## Test model trained on UFDR with CCPD

### Model 4.1: Fine-tune the model on CCPD dataset
